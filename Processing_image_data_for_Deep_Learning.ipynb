**Libraries that can be used for image processing**


1.   matplotlib.image
2.   Pillow
3.   OpenCV(cv2)



```python
# importing the image module from matplotlib library
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# loading an image through matplotlib.image module
img = mpimg.imread('/content/dg.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
print(img)
```

```python
# displaying the image from numpy array
img_plot = plt.imshow(img)
plt.show()
```

Resizing the image using Pillow library

```python
from PIL import Image
```

```python
img1 = Image.open('/content/dg.jpg')
img1_resized = img1.resize((200, 200))
```

```python
img1_resized.save('dog_image_resized.jpg')
```

```python
img2 = mpimg.imread('/content/dog_image_resized.jpg')
# displaying the image from numpy array
img_res_plot = plt.imshow(img2)
plt.show()
```

```python
print(img2.shape)
```

**Converting RGB images to Grayscale image using OpenCV library**

```python
# importing OpenCV library
import cv2
```

```python
img3 = cv2.imread('/content/dg.jpg')
```

```python
type(img)
```

```python
img3.shape
```

```python
grayscale_image = cv2.cvtColor(img3, cv2.COLOR_RGB2GRAY)
```

```python
grayscale_image
```

```python
grayscale_image.shape
```
