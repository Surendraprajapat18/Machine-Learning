```python
# Import required librery
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# DataSet Link : https://bit.ly/3J7G3of
credit_card_data = pd.read_csv("/content/creditcard.csv")
```

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

Under Sampling

Build a sample dataset containing similar distribution of normal transaction and Fraudulent Transaction

Number of Fraudulent Transactions == 492

```python
legit_sample = legit.sample(n=492)
```

Concat two dataset 

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.shape
```

```python
new_dataset.groupby('Class').mean()
```

Spliting data into features & Targets

```python
x = new_dataset.drop(columns = 'Class', axis = 1)
y = new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
print(x_train)
```

```python
print(y_train)
```

Model Evalution

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print("Accuracy of traning data Accuracy: ", training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(y_test, x_test_prediction)
```

```python
print("Accuracy of test data Accuracy: ", test_data_accuracy)
```
