# **GridSearchCV**

GridSearchCV is used for determining the best parameters for our model

```python
# loading the SVC model
model = SVC()
```

```python
# hyperparameters 

parameters = {
    'kernel':['linear','poly','rbf', 'sigmoid'],
    'C':[1, 5, 10, 20]
}
```

```python
# grid Search
classifier = GridSearchCV(model, parameters, cv=5)
```

```python
# fiting the data our mode
classifier.fit(x, y)
```

```python
classifier.cv_results_
```

```python
# best parameters
best_parameters = classifier.best_params_
print(best_parameters)
```

```python
# highest accuraacy 
high_accuracy = classifier.best_score_
print(high_accuracy)
```

```python
 # loading the results to pandas dataframe
results = pd.DataFrame(classifier.cv_results_)
```

```python
grid_search_results = results[['param_C', 'param_kernel', 'mean_test_score']]
```
