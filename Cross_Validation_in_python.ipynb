Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

Importing the models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Dataset link : https://bit.ly/3H2SQWj
heart_data = pd.read_csv("/content/heart.csv")
```

```python
x = heart_data.drop(columns='target', axis=1)
y = heart_data['target']
```

```python
x
```

# **Train Test Split**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=3)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Comparing the perfomance of the models

```python
# list of models
models = [LogisticRegression(max_iter=1000), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
def compare_models_train_test(models):
  for model in models:

    model.fit(x_train, y_train)
    test_data_prediction = model.predict(x_test)
    accuracy = accuracy_score(y_test, test_data_prediction)
    print('Accuracy score of the', model, ' = ', accuracy)
```

```python
compare_models_train_test(models)
```

# **Cross Validation**

Logistic Regression

```python
cv_score_lr = cross_val_score(LogisticRegression(max_iter = 1000), x, y, cv=5) # cv=5 use iteration 5 is default
print(cv_score_lr)
print(round(np.mean(cv_score_lr) * 100, 2))
```

Support Vector Classifier

```python
cv_score_SVC = cross_val_score(SVC(kernel='linear'), x, y, cv=5) # cv=5 use iteration 5 is default
print(cv_score_SVC)
print(round(np.mean(cv_score_SVC) * 100, 2))
```

KNeighbours Classifier

```python
cv_score_K = cross_val_score(KNeighborsClassifier(), x, y, cv=5) # cv=5 use iteration 5 is default
print(cv_score_K)
print(round(np.mean(cv_score_K) * 100, 2))
```

Creating the finction to compare all models

```python
# list of models
models = [LogisticRegression(max_iter=1000), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
def compare_models_cross_validation():

  for model in models:
    
    cv_score = cross_val_score(model, x, y, cv=5)
    mean_accuracy = round(np.mean(cv_score) * 100, 2)
    print("Cross Validation accuracies for the", model, '=', cv_score)
    print('Accuracy % of the', model, mean_accuracy)
    print("---------------------------------------------")
```

```python
compare_models_cross_validation()
```
